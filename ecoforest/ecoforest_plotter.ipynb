```python
import ipywidgets as widgets
from IPython.display import display
from datetime import datetime
date_picker = widgets.DatePicker(
    description='Date',
    disabled=False,
    value=datetime.today()
)
display(date_picker)
```

```python
from main import Dataset
import matplotlib.pyplot as plt

def refresh_day_plot(axes, change):
    date = date_picker.value
    data = Dataset(date.day, date.month, date.year)
    data.plot(axes)
    plt.show()

axes = refresh_day_plot(None, None)
date_picker.observe(lambda x: refresh_day_plot(axes, x), names='value')
```
